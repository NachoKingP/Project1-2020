```python
#Set variables for line plots
all08_x = ALL2008_df["Date"].values.tolist()
all20_x = ALL2020_df["Date"].values.tolist()
amzn08_x = AMZN2008_df["Date"].values.tolist()
amzn20_x = AMZN2020_df["Date"].values.tolist()
c08_x = C2008_df["Date"].values.tolist()
c20_x = C2020_df["Date"].values.tolist()
ccl08_x = CCL2008_df["Date"].values.tolist()
ccl20_x = CCL2020_df["Date"].values.tolist()
cost08_x = COST2008_df["Date"].values.tolist()
cost20_x = COST2020_df["Date"].values.tolist()
tsn08_x = TSN2008_df["Date"].values.tolist()
tsn20_x = TSN2020_df["Date"].values.tolist()

all08_y = ALL2008_df["Close"].values.tolist()
amzn08_y = AMZN2008_df["Close"].values.tolist()
c08_y = C2008_df["Close"].values.tolist()
ccl08_y = CCL2008_df["Close"].values.tolist()
cost08_y = COST2008_df["Close"].values.tolist()
tsn08_y = TSN2008_df["Close"].values.tolist()
all20_y = ALL2020_df["Close"].values.tolist()
amzn20_y = AMZN2020_df["Close"].values.tolist()
c20_y = C2020_df["Close"].values.tolist()
ccl20_y = CCL2020_df["Close"].values.tolist()
cost20_y = COST2020_df["Close"].values.tolist()
tsn20_y = TSN2020_df["Close"].values.tolist()
```

## Generate Weekly X-Axis Values For All Stocks

```python
#Create weekly set of dates for plotting 08 stocks
div = int(round(len(all08_x) / 7,0))
div
x_axis = []
for x in range(div):
    x_axis.append(all08_x[x*7])

#print(x_axis)

#Set 08 x_axes for all stocks
all08_x_short = x_axis
amzn08_x_short = x_axis
c08_x_short = x_axis
ccl08_x_short = x_axis
cost08_x_short = x_axis
tsn08_x_short = x_axis

#Create weekly set of dates for plotting 20 stocks
div = int(round(len(all20_x) / 7,0))
div
x_axis = []
for x in range(div):
    x_axis.append(all20_x[x*7])

#print(x_axis)

#Set 08 x_axes for all stocks
all20_x_short = x_axis
amzn20_x_short = x_axis
c20_x_short = x_axis
ccl20_x_short = x_axis
cost20_x_short = x_axis
tsn20_x_short = x_axis
```

```python
# Plot the Allstate 2008 stock data
ymin = min(all08_y) - min(all08_y)*.01
ymax = max(all08_y) + max(all08_y)*.01

plt.figure(figsize=(20,16))
all08, = plt.plot(all08_x, all08_y, marker=".", color="blue", linewidth=1, label="ALL Closing Price")
plt.legend(handles=[all08], loc="best", fontsize=20)
plt.xlabel("Dates", size=20)
plt.xticks(all08_x_short, rotation=45, size=20)
plt.yticks(size=20)
plt.ylabel("Stock Closing Price", size=20)
plt.ylim(ymin, ymax)
plt.title("Allstate Stock Prices During 2008 Recession", size=20)
plt.savefig("../output/Allstate_2008.png")
plt.show()
```
